# Renaming and Selecting the revelent columns

```python
# Renaming and selecting the relevant columns
data.rename(columns={'iyear':'Year','imonth':'Month','iday':"day",'gname':'Group','country_txt':'Country','region_txt':'Region','provstate':'State','city':'City','latitude':'latitude',
    'longitude':'longitude','summary':'summary','attacktype1_txt':'Attacktype','targtype1_txt':'Targettype','weaptype1_txt':'Weapon','nkill':'kill',
     'nwound':'Wound'},inplace=True)
data = data[['Year','Month','day','Country','State','Region','City','latitude','longitude',"Attacktype",'kill',
               'Wound','target1','summary','Group','Targettype','Weapon','motive']]
```

```python
data
```

# Now fill the missing values in the columns

```python
data['Wound'] = data['Wound'].fillna(0)
data['kill'] = data['kill'].fillna(0)
data['casualities'] = data['kill'] + data['Wound']
```

```python
data.head()
```

# Exploratory Data Analysis (EDA)

```python
plt.subplots(figsize = (15,7))
sns.countplot('Year',data=data)
plt.xticks(rotation=90)
plt.xlabel('Years',fontsize=14)
plt.title('No. of Terrorist Activities by Year',fontsize=20)
plt.show()
```

```python
plt.subplots(figsize=(20,6))
sns.countplot('Month',data=data,order=data['Month'].value_counts().index)
plt.xlabel('Attack Type')
plt.title('No. of Terrorist Activities By Month',fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.barplot(data['Country'].value_counts()[:10].index,data['Country'].value_counts()[:10].values)
plt.title("No. of Terrorist Acttacks by Country",fontsize=20)
plt.xlabel('Countries',fontsize=16)
plt.ylabel('Count',fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.countplot('Attacktype',data=data,order=data['Attacktype'].value_counts().index)
plt.xticks(rotation=15)
plt.title('Terrorist Activities by AttackType',fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.barplot(data['Group'].value_counts()[1:16].values,data['Group'].value_counts()[1:16].index)
plt.title("No. of Terrorist Activities by Terrorist Group",fontsize=20)
plt.show()
```
